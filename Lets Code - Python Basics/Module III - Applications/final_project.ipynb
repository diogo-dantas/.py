```python
def get_datasets(y, labels):    # helpers
    if type(y[0]) == list:
        datasets = []
        for i in range(len(y)):
            datasets.append({
                'label': labels[i],
                'data': y[i]
            })
        return datasets
    else:
        return [
            {
                'labels': labels[0],
                'data': y
            }
        ]
```

```python
def set_title(title=''):        # helpers
    if title != '':
        display = 'true'
    else:
        display = 'false'
    return {
        'title': title,
        'display': display
    }
```

```python
def create_chart(x, y, labels, kind='bar', title=''):    # helpers

    datasets = get_datasets(y, labels)
    options = set_title(title)

    chart = {
        'type': kind,
        'data': {
            'labels': x,
            'datasets': datasets
        },
        'options': options
    }
    return chart
```

```python
def get_api_chart(chart):    # helpers
    url_base = 'https://quickchart.io/chart'
    resp = r.get(f'{url_base}?c={str(chart)}')
    return resp.content  # receive the binary value
```

```python
def save_image(path, content):       # helpers
    with open(path, 'wb') as image:
        image.write(content)
```

```python
from PIL import Image             
from IPython.display import display
```

```python
def display_image(path):           # helpers
    img_pil = Image.open(path)
    display(img_pil)
```

```python
y_data_1 = []

for obs in final_data[1::10]:
    y_data_1.append(obs[CONFIRMED])
    
y_data_2 = []

for obs in final_data[1::10]:
    y_data_2.append(obs[RECOVERED])
    
labels = ['Confirmed','Recovered']

x = []

for obs in final_data[1::10]:
    x.append(obs[DATE].strftime('%d/%m/%Y'))
    
chart = create_chart(x, [y_data_1, y_data_2], labels, title ="Chart confirmed vs recovered")

chart_content = get_api_chart(chart)

save_image('My_first_chart.png', chart_content)
display_image('My_first_chart.png')
```

```python
def get_api_qrcode(link):   
    text = quote(link)  # url link parsing
    url_base = 'https://quickchart.io/qr'
    resp = r.get(f'{url_base}?text={text}')
    return resp.content
```
